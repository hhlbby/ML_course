## Коротко о теории логистической регрессии

### 1. Логистическая регрессия
Логистическая регрессия моделирует вероятность принадлежности объекта к положительному классу ($y=1$):

$$P(y=1 \mid x) = \sigma(z), \quad z = w^T x + b$$



Где $\sigma(z)$ — сигмоида:

$$\sigma(z) = \frac{1}{1 + e^{-z}}$$

### 2. Почему не линейная регрессия?
В линейной регрессии предсказание $z = w^T x + b$ может принимать любые значения от $-\infty$ до $+\infty$. Для классификации нам нужна вероятность в диапазоне $[0, 1]$. Сигмоида служит «активацией», которая сжимает любое вещественное число в этот интервал.

### 3. Функция потерь
Для обучения используется логистическая функция потерь (**Log-Loss** или бинарная кросс-энтропия):

$$L = - \frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(\hat{p}_i) + (1 - y_i) \log(1 - \hat{p}_i) \right]$$

Здесь:
* $y_i$ — истинный класс (0 или 1);
* $\hat{p}_i$ — предсказанная моделью вероятность.

### 4. Обучение
Минимизация **Log-Loss** происходит с помощью **градиентного спуска**. Веса обновляются итеративно в направлении, обратном градиенту:

$$w_{new} = w_{old} - \eta \cdot \nabla_w L$$

Где $\eta$ — скорость обучения (learning rate).

### 5. Метрики для сравнения
После реализации алгоритма «вручную», результат сравнивается с `sklearn.linear_model.LogisticRegression` по следующим показателям:

| Метрика | Описание |
| :--- | :--- |
| **Accuracy** | Доля правильных предсказаний. |
| **Precision** | Способность модели не помечать отрицательный класс как положительный. |
| **Recall** | Способность модели находить все объекты положительного класса. |
| **F1-score** | Гармоническое среднее между Precision и Recall. |
| **ROC-AUC** | Способность модели разделять классы (площадь под кривой). |



### 6. Почему нужна стандартизация
Градиентный спуск чувствителен к масштабу признаков. Если один признак варьируется от 0 до 1, а другой — от 0 до 1 000 000, изолинии функции потерь будут сильно вытянуты, что замедлит сходимость или приведет к расходимости алгоритма. Стандартизация делает «ландшафт» функции потерь более симметричным.

## Шаг 2. Реализуем сигмоиду

Допишите функцию `sigmoid(z)`.

Проверка:
- `sigmoid(0)` должно быть равно `0.5`;
- функция должна работать и со скалярами, и с `numpy`-массивами.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

print("sigmoid(0) =", sigmoid(0))
print("sigmoid([-1, 0, 1]) =", sigmoid(np.array([-1.0, 0.0, 1.0])))
```

## Шаг 3. Реализуем log-loss

Напишите функцию `compute_log_loss(y_true, y_pred_proba)`.

Важно:
- чтобы избежать `log(0)`, используйте `np.clip(y_pred_proba, 1e-15, 1 - 1e-15)`.

```python
def compute_log_loss(y_true, y_pred_proba):
    y_pred_proba = np.clip(y_pred_proba, 1e-15, 1 - 1e-15)
    loss = - np.mean([y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred) for y, y_pred in zip(y_true, y_pred_proba)])
    return loss

toy_y = np.array([0, 1, 1, 0])
toy_p = np.array([0.1, 0.9, 0.8, 0.3])
print("Toy log-loss:", compute_log_loss(toy_y, toy_p))
```

## Шаг 4. Один шаг градиентного спуска

Напомним:
- \(z = Xw + b\)
- \(\hat{p} = \sigma(z)\)

Градиенты для log-loss:
\[
dw = 
rac{1}{n} X^T(\hat{p} - y), \qquad db = 
rac{1}{n}\sum(\hat{p} - y)
\]

Ниже нужно реализовать функцию, которая считает:
- вероятности,
- loss,
- градиенты `dw`, `db`.

```python
def forward_backward(X, y, w, b, lmd):
    z = X @ w + b
    p = sigmoid(z)
    loss = compute_log_loss(y, p) + lmd * np.sum(w ** 2)
    n = len(X)
    
    dw = X.T @ (p - y) / n + lmd * w
    db = np.sum(p - y) / n 
    
    return p, loss, dw, db

n_features = X_train_scaled.shape[1]
w0 = np.zeros(n_features)
b0 = 0.0
lmd0 = 0.001
p0, loss0, dw0, db0 = forward_backward(X_train_scaled, y_train.values, w0, b0, lmd0)
print("Initial loss:", loss0)
print("dw shape:", dw0.shape)
print("db:", db0)
```

## Шаг 5. Собираем модель в класс

Допишите методы:
- `fit`
- `predict_proba`
- `predict`

Требования:
- обучение через градиентный спуск;
- сохранять историю значения loss;
- в `predict` использовать порог `threshold`.

```python
class MyLogisticRegressionGD:
    def __init__(self, learning_rate=0.05, n_iters=3000, threshold=0.5, lmd=0.01):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.threshold = threshold
        self.w = None
        self.b = None
        self.lmd = lmd
        self.loss_history = []

    def fit(self, X, y):
        w = np.zeros(n_features)
        b = 0.
        for _ in range(self.n_iters):
            p, l, dw, db = forward_backward(X, y, w, b, self.lmd)
            self.loss_history.append(l)
            w -= self.learning_rate * dw
            b -= self.learning_rate * db
        self.w = w
        self.b = b
        return self

    def predict_proba(self, X):
        return sigmoid(X @ self.w + self.b)

    def predict(self, X):
        prob = self.predict_proba(X)
        return [(1 if p >= self.threshold else 0) for p in prob]
```

## Шаг 6. Обучаем свою модель

Попробуйте обучить свою модель и посмотрите:
- уменьшается ли `loss`;

```python
my_model = MyLogisticRegressionGD(
    learning_rate=0.002,   # подберите разумное значение
    n_iters=500,         # задайте число итераций
    threshold=0.5,
    lmd=10
)

my_model.fit(X_train_scaled, y_train)
my_proba_train = my_model.predict_proba(X_train_scaled)
my_pred_train = my_model.predict(X_train_scaled)
my_metrics = {
    "accuracy": accuracy_score(y_train, my_pred_train),
    "precision": precision_score(y_train, my_pred_train),
    "recall": recall_score(y_train, my_pred_train),
    "f1": f1_score(y_train, my_pred_train) ,
    "roc_auc": roc_auc_score(y_train, my_proba_train)
}


plt.figure(figsize=(8, 4))
plt.plot(my_model.loss_history)
plt.title("История функции потерь")
plt.xlabel("Итерация")
plt.ylabel("Log-loss")
plt.show()
print(pd.Series(my_metrics).round(4))
```

## Ответы на вопросы

1. Сигмоида принимает значения от 0 до 1, поэтому ее интерпретируют как вероятность попадания объекта в той или иной класс
2. log-loss основана на ММП, что является основой для логистической регрессии. MSE больше подходит для линейной регрессии
3. Градиентный спуск чувствителен к масштабу признаков, если один признак варьируется от 0 до 1, а другой - от 0 до 10000, это замедлит обучение или приведет к расходимости алгоритма
4. Достаточно близки, но все равно уступают scikit-learn
5. Библиотечная реализация использует другие оптимизирующие фичи, например, использует не только градиент, но и следит за кривизной функции потерь
6. precision уменьшится, а recall увеличится
7. В задачах, где данные разделены линейно, логистическая регрессия очень хороша, но в ином случае лучше использовать иные методы
